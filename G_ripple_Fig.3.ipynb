```python
data12= pd.read_pickle('G_ripple_data12.pkl')
```

```python
data12.head(1)
```

```python
data12_x = data12.groupby(['strength_interval','group'], as_index=False).agg({"delta_C_ITE":np.nanmean})
```

```python
data12_x.head()
```

```python
data12_x['strength_interval'] = data12_x['strength_interval'].astype(str)
data12_x['strength_interval'] = data12_x['strength_interval'].str.strip('[]()')
data12_x[['interval_min', 'interval_max']] = data12_x['strength_interval'].str.split(', ', expand=True)
data12_x['interval_min'] = pd.to_numeric(data12_x['interval_min'])
data12_x['interval_max'] = pd.to_numeric(data12_x['interval_max'])
data12_x['strength'] = (data12_x['interval_min'] + data12_x['interval_max']) / 2
```

```python
data12_x.head()
```

```python
sns.set(font_scale=1.5)
sns.set(style="ticks")
color_mapping = {'coauthor': '#33A02C'}
sns.lmplot(x='strength', y='delta_C_ITE', data=data12_x,ci=95,hue='group',order=1,palette=color_mapping,legend=False,scatter_kws={"alpha": 0.7, "s": 60}, line_kws={"linewidth": 6},height=7,aspect=1.67)
plt.xticks(fontsize=33)
plt.yticks(fontsize=33)
plt.ylim(-0.06, 0.078)
plt.xlabel('Collaboration strength',fontsize=50)
plt.ylabel(r"ΔC$^{ITE}_{c}$",fontsize=50)
plt.show()
```
